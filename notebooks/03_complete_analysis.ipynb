This notebook performs end-to-end analysis and generates all portfolio outputs.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import sys
import os

# Add src to path
sys.path.append('..')

# Import our modules
from src.data_loader import load_world_bank_data
from src.data_cleaner import clean_economic_data

# Create output directories
os.makedirs('../outputs/figures', exist_ok=True)
os.makedirs('../outputs/dashboards', exist_ok=True)
os.makedirs('../outputs/reports', exist_ok=True)

print('✅ Setup complete')
```

```python
# Load and clean data
df_raw = load_world_bank_data()
df = clean_economic_data(df_raw)

# Save cleaned data
df.to_csv('../outputs/cleaned_data.csv', index=False)
print("✅ Data ready. Shape:", df.shape)
df.head()
```

```python
# Generate summary statistics
summary = df.describe()
summary.to_csv('../outputs/reports/summary_statistics.csv')

country_stats = df.groupby('country')['gdp_growth'].agg(['mean', 'std']).round(2)
country_stats.to_csv('../outputs/reports/country_rankings.csv')

print("📈 Summary Statistics:")
display(summary)
```

```python
# Create visualizations
plt.style.use('seaborn-v0_8-darkgrid')

# GDP Trends
plt.figure(figsize=(12, 6))
top_5 = df.groupby('country')['gdp_growth'].mean().nlargest(5).index
for country in top_5:
    data = df[df['country'] == country]
    plt.plot(data['year'], data['gdp_growth'], marker='o', label=country)
plt.title('GDP Growth Trends - Top 5 Countries')
plt.xlabel('Year')
plt.ylabel('GDP Growth (%)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('../outputs/figures/gdp_trends.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Interactive dashboard
fig = px.line(df, x='year', y='gdp_growth', color='country',
              title='Interactive GDP Growth Dashboard')
fig.write_html('../outputs/dashboards/interactive_dashboard.html')
fig.show()
```
